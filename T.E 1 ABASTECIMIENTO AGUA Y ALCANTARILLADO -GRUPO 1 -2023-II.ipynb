El objetivo del sistema de purificación de agua es suministrar agua potable de alta calidad de manera eficaz a la población. De acuerdo con investigaciones previas, se ha concluido que el agua procedente de las áreas de recolección no requiere un proceso de purificación altamente complicado, ya que se trata de agua natural de manantial. Por lo tanto, se propone la instalación de un sistema de purificación de agua en el punto de recolección, el cual estará ubicado antes del depósito y estará compuesto por la zona de recolección, una cámara de mezcla, un floculador, un sedimentador y un filtro. </p>

 <img src="FIGURA 13-DESCRIPCION.png" > </p>

Remover compuestos orgánicos y sustancias volátiles productoras de olores y sabores provenientes normalmente de aguas de manantial, además de proporcionarle a los microorganismos el oxígeno necesario para que realicen sus procesos de transformación y degradación de la materia orgánica contaminante.</p>

**e.	Sedimentador**</p>
El principal propósito de un sedimentador es facilitar la sedimentación al reducir la turbulencia y la velocidad de la corriente de aguas residuales</p>
**f.	Filtro**</p>
El mecanismo biológico de esta filtración es muy efectivo para la eliminación de micro contaminantes. Se produce una reducción de la turbidez inicial del agua y de los coliformes de hasta un 90 o 99%.</p>

Es un evento ya sea gradual o repentino, que puede ser anticipado con los medios técnicos adecuados y gestionado mediante la implementación de infraestructuras adecuadas. Estos eventos pueden ocurrir en cursos de agua como ríos y en áreas costeras. En la región, las inundaciones están vinculadas a fenómenos como tsunamis y mareas altas en las zonas costeras, mientras que tierra adentro, son el resultado de intensas precipitaciones y la expansión de las planicies de inundación de los ríos. En terrenos con pendientes pronunciadas, es posible que se produzcan crecidas repentinas de aparición rápida y desaparición veloz.</p>
La gravedad de los impactos está relacionada con varios factores, como el tamaño del área inundada, la profundidad del agua, la velocidad de la corriente, la cantidad de sedimentos transportados, la duración y la frecuencia de las inundaciones. En las llanuras, las inundaciones pueden alterar el curso de los ríos y depositar sedimentos. Estas crecidas pueden erosionar el lecho del río y generar nuevos deslizamientos o reactivar los antiguos.</p>
Las consecuencias de las inundaciones y las crecidas en los sistemas incluyen:</p>
* La destrucción total o parcial de estructuras de captación ubicadas en ríos o arroyos. </p>
* La acumulación de sedimentos que obstruyen componentes de infraestructura.</p>
* La pérdida de instalaciones de captación debido al cambio en el cauce del río. </p>
* La rotura de tuberías de distribución y conexiones en áreas costeras, a menudo a causa del impacto de las mareas altas y la proximidad a los cursos de agua. </p>
* La contaminación del agua en las cuencas. El daño a los equipos de bombeo al estar en contacto con el agua.</p>
* Adicionalmente, estos eventos pueden tener efectos indirectos, como la interrupción del suministro de energía eléctrica, el cierre de carreteras y la interrupción de las comunicaciones.</p>
**Sequías**</p>
Es una condición gradual que se desarrolla lentamente a lo largo de varios años, siendo predecible cuando se dispone de las herramientas técnicas apropiadas y manejable a largo plazo mediante la implementación de medidas adecuadas. La magnitud del impacto está estrechamente vinculada a factores como la escasez de lluvias, el volumen de precipitaciones, la duración de la sequía, la erosión del terreno y la extensión de la zona climática desértica.</p>
La sequía provoca cambios significativos en el entorno bioclimático de la región y en las condiciones del agua subterránea. Los efectos de las sequías en los sistemas incluyen:</p>
* La pérdida o reducción del caudal de agua en cuerpos de agua superficiales y/o subterráneos.</p> 
* La implementación de restricciones y la posibilidad de suspender el suministro de servicios.</p>

**Deslizamientos**</p>
Es un evento que puede manifestarse de manera gradual o repentina, y en algunas ocasiones es predecible, controlable y modificable. Las fallas súbitas del terreno pueden suceder sin previo aviso, mientras que las fallas de evolución lenta presentan señales precursoras que pueden ser identificadas y monitoreadas con la instrumentación adecuada.</p>
La magnitud del impacto está relacionada con factores como el volumen de material desplazado, la velocidad y dirección de la masa en movimiento, el tamaño de las rocas involucradas y la naturaleza del desplazamiento, todo ello influenciado por la ubicación geográfica del sistema.</p>
Tanto los macro deslizamientos como los desplazamientos de terreno inducidos por sismos o precipitaciones pueden tener un impacto significativo en la topografía local de la zona. Los deslizamientos provocan efectos directos debido a la deformación y la colisión de la masa en movimiento, así como efectos secundarios relacionados con las inundaciones aguas arriba de un deslizamiento o represamiento, así como las crecidas aguas abajo tras la rotura de dicho deslizamiento o represamiento.</p>
**Los efectos de los deslizamientos en los sistemas son:**</p>
* La destrucción total o parcial de todas las estructuras, especialmente las de captación y conducción, ubicadas sobre la trayectoria principal de los deslizamientos activos, especialmente en terrenos montañosos inestables con fuertes pendientes o en pendientes muy pronunciadas susceptibles a deslizamientos.</p>
* La contaminación del agua en áreas de captación superficial en regiones montañosas, lo que puede tener efectos indirectos como la interrupción del suministro eléctrico, la obstrucción de vías de acceso y las comunicaciones.</p>
**Sismos o Terremotos**</p>
Evento súbito, no predecible, no controlable ni alterable por el hombre. La gravedad del impacto se relaciona con la magnitud de la energía liberada, la distancia y ubicación del epicentro del terremoto en relación con el elemento expuesto y las condiciones locales del terreno.</p>
El tamaño del área afectada está directamente relacionado con la cantidad de energía liberada por el evento e inversamente con la profundidad del sitio de liberación de energía. El terremoto es capaz de modificar y destruir el entorno físico de la región.</p>
Un terremoto tiene efectos directos y secundarios. Los efectos directos son aquellos causados por el sacudimiento producido por el paso de la onda sísmica y los secundarios por las deformaciones permanentes del terreno, como: asentamientos diferenciales del suelo, deslizamientos y correntadas de lodo, licuación del suelo, avalanchas, maremotos o tsunamis.</p>
Los efectos del sismo en los sistemas de agua potable son:</p>
* Destrucción total o parcial de las estructuras de captación, conducción, tratamiento, almacenamiento y distribución. </p>
* Roturas de las tuberías de conducción y distribución y daños en las uniones, entre tuberías o con los tanques, con la consiguiente pérdida de agua.</p>
*  Interrupción de la corriente eléctrica, de las comunicaciones y de las vías de acceso.Modificaciones de la calidad del agua por deslizamientos en áreas de topografía montañosa. </p>
* Variación(disminución) del caudal en captaciones subterráneas o superficiales. </p>
* Cambio del sitio de salida del agua en manantiales.</p>

**Volumen de almacenamiento**</p>
El volumen de almacenamiento está conformado por el volumen de regulación, volumen contra incendio y volumen de reserva.  
**Volumen de regulación** </p>
La norma OS. 030 de almacenamiento de agua para consumo humano indica que si no se encuentra con información de las variaciones horarias de la demanda se debe adoptar como mínimo el 25% del promedio anual de la demanda como capacidad de regulación, siempre que el suministro de la fuente de abastecimiento sea calculado para 24 horas de funcionamiento. Entonces el volumen de regulación para este proyecto será el siguiente. </p>
Qmd=0.37 Lt/seg</p>
Vm=Qmdx 86400/1000 m^3→Vm=6.38x 86400/1000 m^3=32m^3</p>
Vr=25%(Vm)=800%m^3=8m^3</p>
**Vr=8m^3**</p>

**Volumen contra incendio**</p>
La norma indica que no se considera obligatorio la demanda contra incendio cuando la población de diseño es menor a 10000 habitantes. La población de diseño para este proyecto es 135 por lo tanto no se considerará el volumen contra incendio. </p>
**Vincendios=0.00 m^3**</p>
**Volumen de Reserva**</p>
El volumen de reserva para este proyecto será el 25% del volumen total calculado, en este caso es el 25% del volumen de regulación determinado ya que el volumen contra incendios es 0, entonces ser tiene: </p>
 Vreserva=25%(Vr+Vincendios)→Vreserva=25%(8) m^3</p>
**Vreserva=2m^3**</p>
**Volumen total del reservorio**</p>
Como se indico anteriormente este es la suma de los volúmenes de regulación, contra incendio y de reserva. Entonces se tiene el siguiente volumen: 
Vtotal=Vregulación+Vincendios+Vreserva</p>
Vtotal=8+0+2=10 m^3</p>
**Vtotal=10m^3**</p>
Entonces según este resultado consideraremos un volumen de diseño igual a 
10 m^3 para el reservorio. </p>
**V(a diseñar )=10m^3**</p>
